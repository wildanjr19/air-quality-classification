```python
import pandas as pd
import matplotlib.pyplot as plt

# load file
df = pd.read_csv("../result/history_v6.csv")
df.head()
```

```python
# hyperparameter untuk visualisasi
title_fontsize = 16         
label_fontsize = 14      
legend_fontsize = 12     
tick_fontsize = 11       
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 8))

# Grafik Loss
ax1.plot(df['epoch'], df['loss'], label='Training Loss', color='red', linewidth=2, marker='o')
ax1.plot(df['epoch'], df['val_loss'], label='Validation Loss', color='orange', linewidth=2, marker='o')
ax1.set_xlabel('Epoch', fontsize=label_fontsize)
ax1.set_ylabel('Loss', fontsize=label_fontsize)
ax1.set_title('Training & Validation Loss', fontsize=title_fontsize)
ax1.legend(fontsize=legend_fontsize)
ax1.grid(True, alpha=0.3)
ax1.tick_params(axis='both', which='major', labelsize=tick_fontsize)

# Grafik Accuracy
ax2.plot(df['epoch'], df['accuracy'], label='Training Accuracy', color='blue', linewidth=2, marker='o')
ax2.plot(df['epoch'], df['val_accuracy'], label='Validation Accuracy', color='green', linewidth=2, marker='o')
ax2.set_xlabel('Epoch', fontsize=label_fontsize)
ax2.set_ylabel('Accuracy', fontsize=label_fontsize)
ax2.set_title('Training & Validation Accuracy', fontsize=title_fontsize)
ax2.legend(fontsize=legend_fontsize)
ax2.grid(True, alpha=0.3)
ax2.tick_params(axis='both', which='major', labelsize=tick_fontsize)

plt.tight_layout()
plt.show()
```
