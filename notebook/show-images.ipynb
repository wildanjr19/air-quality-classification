```python
import os
from PIL import Image

import matplotlib.pyplot as plt

# Path to the training data folder
train_dir = '../data/train'

# Get class names (subfolder names)
class_names = sorted(os.listdir(train_dir))

# Prepare figure
fig, axes = plt.subplots(2, 2, figsize=(8, 8))
axes = axes.flatten()

# Show one image per class
for idx, class_name in enumerate(class_names[:4]):
    class_folder = os.path.join(train_dir, class_name)
    img_name = os.listdir(class_folder)[0]  # Take the first image
    img_path = os.path.join(class_folder, img_name)
    img = Image.open(img_path)
    axes[idx].imshow(img)
    axes[idx].set_title(class_name)
    axes[idx].axis('off')

plt.tight_layout()
plt.show()
```

```python
# Ambil nama kelas dari folder train dan val
val_dir = '../data/val'
all_dirs = [train_dir, val_dir]

# Hitung total gambar per kelas (gabungan train dan val)
class_counts = {}
for dir_path in all_dirs:
    for class_name in class_names:
        class_folder = os.path.join(dir_path, class_name)
        if os.path.exists(class_folder):
            count = len(os.listdir(class_folder))
            class_counts[class_name] = class_counts.get(class_name, 0) + count

# Plot grafik batang
plt.figure(figsize=(6,4))
plt.bar(class_counts.keys(), class_counts.values(), color='skyblue')
plt.xlabel('Kelas')
plt.ylabel('Jumlah Gambar')
plt.title('Distribusi Jumlah Gambar')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```
